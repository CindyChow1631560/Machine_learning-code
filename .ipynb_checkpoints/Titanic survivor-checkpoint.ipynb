```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import ShuffleSplit,GridSearchCV,StratifiedShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFECV
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

read data from files

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
#description about train data
train.describe()
```

```python
#Survival vs gender indicates that females have more chances to survive
gender_pivot=train.pivot_table(index="Sex",values="Survived")
gender_pivot.plot.bar()
plt.show()
```

```python
#Survival vs classes indicates that higher classes have more chances to survive
class_pivot=train.pivot_table(index="Pclass",values="Survived")
class_pivot.plot.bar()
plt.show()
```

```python
##suivival vs family size indicates that 
family=train[['SibSp','Parch','Survived']]
family_size=family[['SibSp','Parch']].sum(axis=1)
family['family_size']=family[['SibSp','Parch']].sum(axis=1)
family['isAlone']=np.where(family_size>=1,1,0)
family_pivot=family.pivot_table(index='family_size',values='Survived')
isAlone_pivot=family.pivot_table(index='isAlone',values='Survived')
family_pivot.plot.bar()
isAlone_pivot.plot.bar()
plt.show()
```

```python
#data preprocessing
train["Cabin_type"]=train["Cabin"].str[0]
train["Fare"]=train["Fare"].fillna(train["Fare"].mean())
train["Embarked"]=train["Embarked"].fillna("S")
train["Cabin_type"]=train["Cabin_type"].fillna("Unknown")
train["Age"]=train["Age"].fillna(-0.5)

test["Cabin_type"]=test["Cabin"].str[0]
test["Fare"]=test["Fare"].fillna(test["Fare"].mean())
test["Embarked"]=test["Embarked"].fillna("S")
test["Age"]=test["Age"].fillna(-0.5)
test["Cabin_type"]=test["Cabin_type"].fillna("Unknown")
```

```python
#categorize Age
bins=[-1,0,5,12,18,35,60,100]
labels=["missing","infant","child","teenager","young_adult","audlt","elder"]
train["ageCategory"]=pd.cut(train["Age"],bins,labels=labels)
test["ageCategory"]=pd.cut(test["Age"],bins,labels=labels)
```

```python
#categorize Fare
fare_cuts = [-1,15,32,100,1000]
fare_labels = ["0-15","15-32","32-100","100+"]
train["Fare_categories"] = pd.cut(train["Fare"],fare_cuts,labels=fare_labels)
test["Fare_categories"] = pd.cut(test["Fare"],fare_cuts,labels=fare_labels)
```

```python
#title feature
titles = {
        "Mr" :         "Mr",
        "Mme":         "Mrs",
        "Ms":          "Mrs",
        "Mrs" :        "Mrs",
        "Master" :     "Master",
        "Mlle":        "Miss",
        "Miss" :       "Miss",
        "Capt":        "Officer",
        "Col":         "Officer",
        "Major":       "Officer",
        "Dr":          "Officer",
        "Rev":         "Officer",
        "Jonkheer":    "Royalty",
        "Don":         "Royalty",
        "Sir" :        "Royalty",
        "Countess":    "Royalty",
        "Dona":        "Royalty",
        "Lady" :       "Royalty"
    }
train_titles= train["Name"].str.extract(' ([A-Za-z]+)\.',expand=False)
train["titles"]=train_titles.map(titles)

familysize_train=train[["SibSp","Parch"]].sum(axis=1)
train['isAlone']=np.where(familysize_train>=1,1,0)

test_titles=train["Name"].str.extract('([A-Za-z]+)\.',expand=False)
test["titles"]=test_titles.map(titles)
familysize_test=test[["SibSp","Parch"]].sum(axis=1)
test['isAlone']=np.where(familysize_test>=1,1,0)
```

```python
columNames=["ageCategory","Sex","titles","Pclass","Fare_categories","Cabin_type","Embarked"]
for colums in columNames:
    train_dummies=pd.get_dummies(train[colums],prefix=colums)
    train=pd.concat([train,train_dummies],axis=1)
    
    test_dummies=pd.get_dummies(test[colums],prefix=colums)
    test=pd.concat([test,test_dummies],axis=1)
    
train.head(5)
```

```python
columns=['PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch',
       'Ticket', 'Fare', 'Cabin', 'Embarked', 'Cabin_type', 'ageCategory',
       'Fare_categories', 'titles', 'isAlone', 'ageCategory_audlt',
       'ageCategory_child', 'ageCategory_elder', 'ageCategory_infant',
       'ageCategory_missing', 'ageCategory_teenager',
       'ageCategory_young_adult', 'Sex_female', 'Sex_male', 'titles_Master',
       'titles_Miss', 'titles_Mr', 'titles_Mrs', 'titles_Officer',
       'titles_Royalty', 'Pclass_1', 'Pclass_2', 'Pclass_3',
       'Fare_categories_0-15', 'Fare_categories_100+', 'Fare_categories_15-32',
       'Fare_categories_32-100', 'Cabin_type_A', 'Cabin_type_B',
       'Cabin_type_C', 'Cabin_type_D', 'Cabin_type_E', 'Cabin_type_F',
       'Cabin_type_G', 'Cabin_type_Unknown', 'Embarked_C', 'Embarked_Q',
       'Embarked_S']
#model selection
def get_model(df,features):
    train_x=df[features]
    train_y=df["Survived"]
    cv=ShuffleSplit(n_splits=10,test_size=0.3,random_state=0)
    model_params=[
       {
            "name": "RandomForestClassifier",
            "estimator": RandomForestClassifier(random_state=0),
            "hyperparameters":
            {
                "n_estimators": [20,25, 35, 40,45, 50, 55, 60, 65, 70, 75],
                "criterion": ["entropy", "gini"],
                "max_features": ["log2", "sqrt"],
                "min_samples_leaf": [1, 5, 8],
                "min_samples_split": [2, 3, 5]    
            } 
       },
        {
            "name": "DecisionTreeClassifier",
            "estimator": DecisionTreeClassifier(),
            "hyperparameters":
            {
               "criterion": ["entropy", "gini"],
                "max_depth": [None,2,4,6,8,10,12,14],
                "min_samples_split": [2,4,6,8,10,12,14],
                "min_samples_leaf": [1,2,3,4,5,6,7,8],
                "max_features": [None,"auto","log2"]
            }
        },
        {
            "name": "KNeighborsClassifier",
            "estimator": KNeighborsClassifier(),
            "hyperparameters":
            {
                "n_neighbors": np.arange(1,10,2),
                "weighs": ["uniform","distance"],
                "algoritnm": ["ball_tree","kd_tree","brute","auto"],
                "p": [1,2,3]
            }
            
        },
        {
            "name": "KernelSVMClassifier",
            "estimator": SVC(random_state=0),
            "hyperparameters":
            {
                "kernel": ["rbf"],
                "C": np.logspace(-9, 3, 13),
                "gamma": np.logspace(-9, 3, 13)
            }
        } ,
        {
            "name": "LogisticRegressionClassifier",
            "estimator": LogisticRegression(),
            "hyperparameters":
            {
                "solver": ["newton-cg", "lbfgs", "liblinear"]
            }
        }
    ]
    models=[]
    for model in model_params:
        print(model["name"])
        grid=GridSearchCV(model["estimator"],param_grid=model["hyperparameters"],cv=10)
        grid.fit(train_x,train_y)
        model_att={
            "model": grid.best_estimator_ ,
            "best_param": grid.best_params_ ,
            "best_score": grid.best_score_,
            "grid": grid
        }
        models.append(model_att)
        print("model and its parameters:")
        print(grid.best_params_)
        print(grid.best_score_)
    return models
```
